```python
#Radio de giro a 298K

import numpy as np
import matplotlib.pyplot as plt

def read_xvg(path):
    data = np.loadtxt(path, comments=["@", "#"])
    return data[:,0], data[:,1]

#Ruta
archivo_298K = "/home/alumno23/MM/for-students/TAREA_PERFECTA/4-analysis/analysis_298K/gyrate.xvg"

#Leer datos
t1, g1 = read_xvg(archivo_298K)

#Plot
plt.plot(t1, g1, label="298K")

plt.xlabel("Tiempo (ps)")
plt.ylabel("Radio de giro (nm)")
plt.title("Radio de giro 298K", fontweight='bold')
plt.legend()
plt.grid()

plt.tight_layout()
plt.show()
```

```python
#Gráfica conjunta: 280K vs 400K

import numpy as np
import matplotlib.pyplot as plt

def read_xvg(path):
    data = np.loadtxt(path, comments=["@", "#"])
    return data[:,0], data[:,1]

#Rutas
archivo_298K = "/home/alumno23/MM/for-students/TAREA_PERFECTA/4-analysis/analysis_298K/gyrate.xvg"
archivo_400K = "/home/alumno23/MM/for-students/TAREA_PERFECTA/4-analysis/analysis_400K/gyrate.xvg"

#Leer datos
t1, g1 = read_xvg(archivo_298K)
t2, g2 = read_xvg(archivo_400K)

#Plot
plt.plot(t1, g1, label="298K")
plt.plot(t2, g2, label="400K")

plt.xlabel("Tiempo (ps)")
plt.ylabel("Radio de giro (nm)")
plt.title("Radio de giro: 298K vs 400K", fontweight='bold')
plt.legend()
plt.grid()

plt.tight_layout()

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

phi_archivo = "/home/alumno23/MM/for-students/TAREA_PERFECTA/4-analysis/analysis_298K/phi-arg-3.dat"
psi_archivo = "/home/alumno23/MM/for-students/TAREA_PERFECTA/4-analysis/analysis_298K/psi-arg-3.dat"

#Cargar datos
phi = np.loadtxt(phi_archivo, usecols=[1])
psi = np.loadtxt(psi_archivo, usecols=[1])
time = np.arange(len(phi))

#Función para graficar evolución temporal
def plot_angle(data, label, color="blue", filename="angle.png"):
    plt.figure(figsize=(8, 5))
    plt.plot(time, data, marker="o", linestyle="-", markersize=2, color=color, label=label)
    plt.xlabel("Paso de tiempo (ps)")
    plt.ylabel(f"Ángulo {label} (grados)")
    plt.title(f"Evolución del ángulo {label} de ARG-3 (298K)", fontweight='bold')
    plt.grid()
    plt.legend()
    
    plt.show()

#Graficar φ y ψ
plot_angle(phi, "φ", "green", "phi_evolucion.png")
plot_angle(psi, "ψ", "purple", "psi_evolucion.png")

# Mapa de Ramachandran
plt.figure(figsize=(7, 7))

# Líneas 0,0 gorditas (centrales)
plt.axhline(0, color='black', linewidth=1.5, zorder=1)
plt.axvline(0, color='black', linewidth=1.5, zorder=1)

plt.scatter(phi, psi, c="blue", alpha=0.5, s=5)
plt.xlabel("Ángulo φ (º)")
plt.ylabel("Ángulo ψ (º)")
plt.title("Mapa de Ramachandran de ARG-3 (298K)", fontweight='bold')
plt.xlim(-180, 180)
plt.ylim(-180, 180)
plt.grid()

plt.show()
```

```python
#Mapa de Ramachandran con visualización mejorada
plt.figure(figsize=(7,7))

#Añadir jitter para que no se vea tan compacto
jitter = 0.5  # grados
plt.scatter(phi + np.random.uniform(-jitter,jitter,len(phi)),
            psi + np.random.uniform(-jitter,jitter,len(psi)),
            c="blue", alpha=0.6, s=15)  # puntos más grandes, alpha → transparencia

plt.xlabel("Ángulo φ (º)")
plt.ylabel("Ángulo ψ (º)")
plt.title("Mapa de Ramachandran de ARG-3 (298K)", fontweight='bold')

#Ajustar límites para que ocupe más espacio
plt.xlim(min(phi)-5, max(phi)+5)
plt.ylim(min(psi)-5, max(psi)+5)

plt.grid()

plt.show()
```

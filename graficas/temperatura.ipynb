```python
import numpy as np
import matplotlib.pyplot as plt

archivos = [
    "/home/alumno23/MM/for-students/TAREA_PERFECTA/4-analysis/analysis_298K/temp.xvg",
    "/home/alumno23/MM/for-students/TAREA_PERFECTA/4-analysis/analysis_400K/temp.xvg",
    "/home/alumno23/MM/for-students/TAREA_PERFECTA/extended/temp.xvg"
]

#Colores gráfico
colores_linea = {
    "analysis_298K": "blue",
    "analysis_400K": "orange",
    "extended": "tab:blue"  
}

for path in archivos:
    try:
        t, temp = np.loadtxt(path, comments=('@', '#'), unpack=True)
        nombre = path.split('/')[-2]
        
        # Seleccionar color según el nombre (si no está en el diccionario usa gris)
        mi_color = colores_linea.get(nombre, "gray")

        #Gráfica de evolución
        plt.figure(figsize=(10, 4))
        plt.plot(t, temp, color=mi_color, lw=1.5, label=f'Temp {nombre}')
        plt.title(f'Evolución Temporal: {nombre}', fontweight='bold')
        plt.xlabel('Tiempo (ps)'); plt.ylabel('Temperatura (K)')
        plt.grid(alpha=0.3)
        plt.legend()
        plt.show()

        #Histograma 
        plt.figure(figsize=(10, 4))
        plt.hist(temp, bins=30, color=mi_color, alpha=0.6, edgecolor='black')
        plt.title(f'Distribución de Temperatura: {nombre}', fontweight='bold')
        plt.xlabel('Temperatura (K)'); plt.ylabel('Frecuencia')
        plt.grid(axis='y', alpha=0.3)
        plt.show()

    except Exception as e:
        print(f"Error en {path}: {e}")
```

```python
import numpy as np
import matplotlib.pyplot as plt

ruta_298 = "/home/alumno23/MM/for-students/TAREA_PERFECTA/4-analysis/analysis_298K/temp.xvg"
ruta_400 = "/home/alumno23/MM/for-students/TAREA_PERFECTA/4-analysis/analysis_400K/temp.xvg"

try:
    #Carga datos
    t1, temp1 = np.loadtxt(ruta_298, comments=('@', '#'), unpack=True)
    t2, temp2 = np.loadtxt(ruta_400, comments=('@', '#'), unpack=True)

    plt.figure(figsize=(12, 6))

    plt.plot(t1, temp1, color='blue', lw=1.2, label='Simulación 298K', alpha=0.8)
    plt.plot(t2, temp2, color='orange', lw=1.2, label='Simulación 400K', alpha=0.8)
    plt.title('Comparativa de Evolución Temporal de la Temperatura', fontsize=14, fontweight='bold')
    plt.xlabel('Tiempo (ps)', fontsize=12)
    plt.ylabel('Temperatura (K)', fontsize=12)
    
    #Línea horizontal para marcar los objetivos teóricos
    plt.axhline(y=298, color='blue', linestyle='--', alpha=0.5, lw=1)
    plt.axhline(y=400, color='orange', linestyle='--', alpha=0.5, lw=1)

    plt.grid(True, alpha=0.3, linestyle='--')
    plt.legend(loc='center right') # Colocamos la leyenda donde menos moleste
    
    plt.tight_layout()

    plt.show()

except Exception as e:
    print(f"Error al generar la comparativa: {e}")
```

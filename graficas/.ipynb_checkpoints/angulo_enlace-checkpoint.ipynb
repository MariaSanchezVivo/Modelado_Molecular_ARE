```python
#[CA-C-O-arg-3] a 298K

import numpy as np
import matplotlib.pyplot as plt

def read_xvg(path):
    data = np.loadtxt(path, comments=["@", "#"])
    return data[:,0], data[:,1]

#Rutas completas
archivo_298K = "/home/alumno23/MM/for-students/TAREA/4-analysis/analysis_298K/angaver.xvg"

#Leer datos
t1, a1 = read_xvg(archivo_298K)

#Configuración del Plot
plt.figure(figsize=(10, 5))

plt.plot(t1, a1, label="298K", color="blue", linewidth=2.0)

plt.xlabel("Tiempo (ps)")
plt.ylabel("Ángulo (°)")
plt.title("Ángulo vs tiempo (ARG-3:CA-C-O)", fontweight='bold')

plt.legend()
plt.grid(True, alpha=0.5)

plt.tight_layout()
plt.show()
```

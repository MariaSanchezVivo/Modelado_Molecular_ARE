```python
import numpy as np
import matplotlib.pyplot as plt

file_298 = "/home/alumno23/MM/for-students/TAREA/4-analysis/analysis_298K/energy.xvg"
file_400 = "/home/alumno23/MM/for-students/TAREA/4-analysis/analysis_400K/energy.xvg"

#Carga de datos
t_298, e_298 = np.loadtxt(file_298, comments=('@', '#'), unpack=True)
t_400, e_400 = np.loadtxt(file_400, comments=('@', '#'), unpack=True)

#Función para gráficas individuales 
def plot_energy_individual(time, energy, label, color, title):
    plt.figure(figsize=(9, 5))
    plt.plot(time, energy, label=label, color=color, lw=1.5)
    
    plt.title(title, fontweight='bold', fontsize=14)
    plt.xlabel("Tiempo (ps)")
    plt.ylabel("Energía Total (kJ/mol)")
    
    plt.legend() 
    plt.grid(alpha=0.3)
    plt.tight_layout()
    plt.show()

#Gráficas
plot_energy_individual(t_298, e_298, "298 K", "blue", "Energía Total a 298 K")
plot_energy_individual(t_400, e_400, "400 K", "orange", "Energía Total a 400 K")

#Gráfica comparativa
plt.figure(figsize=(10, 6))

plt.plot(t_298, e_298, label="298 K", color="blue", lw=1.5, alpha=0.7)
plt.plot(t_400, e_400, label="400 K", color="orange", lw=1.5, alpha=0.7)
plt.title("Comparación de Energía Total", fontweight='bold', fontsize=14)
plt.xlabel("Tiempo (ps)")
plt.ylabel("Energía Total (kJ/mol)")

plt.legend(loc='center right', frameon=True, shadow=True)

plt.grid(alpha=0.3, linestyle='--')
plt.tight_layout()

plt.show()
```

```python
#[CO-arg-3] a 298K
import numpy as np
import matplotlib.pyplot as plt

def read_xvg(path):
    data = np.loadtxt(path, comments=["@", "#"])
    return data[:,0], data[:,1]

#Ruta 
archivo_dist = "/home/alumno23/MM/for-students/TAREA/4-analysis/analysis_298K/dist.xvg"

#Leer datos
t2, d2 = read_xvg(archivo_dist)

#Configuración de la figura para que sea más legible
plt.figure(figsize=(10, 6))

#Dibujar con linewidth más ancho
plt.plot(t2, d2, label="298K", color="blue", linewidth=2.0)

plt.xlabel("Tiempo (ps)", fontsize=12)
plt.ylabel("Distancia (nm)", fontsize=12)
plt.title("Distancia vs tiempo", fontsize=14, fontweight='bold')

plt.legend()
plt.grid(True, linestyle='--', alpha=0.7) # Rejilla más suave

plt.tight_layout()
plt.show()
```

```python
#[CO-arg-3] 298K vs 400K

import numpy as np
import matplotlib.pyplot as plt

def read_xvg(path):
    data = np.loadtxt(path, comments=["@", "#"])
    return data[:,0], data[:,1]

#Rutas 
archivo_298K = "/home/alumno23/MM/for-students/TAREA/4-analysis/analysis_298K/dist.xvg"
archivo_400K = "/home/alumno23/MM/for-students/TAREA/4-analysis/analysis_400K/dist.xvg"

#Leer datos
t1, d1 = read_xvg(archivo_298K)
t2, d2 = read_xvg(archivo_400K)

#Configuración del Plot
plt.figure(figsize=(10, 6))

plt.plot(t1, d1, label="298K", color="blue", linewidth=1.8, alpha=0.8)
plt.plot(t2, d2, label="400K", color="orange", linewidth=1.8, alpha=0.8)

plt.xlabel("Tiempo (ps)", fontsize=12)
plt.ylabel("Distancia (nm)", fontsize=12)
plt.title("Comparativa de Distancia: 298K vs 400K", fontsize=14, fontweight='bold')

plt.legend(loc='upper right') 
plt.grid(True, linestyle='--', alpha=0.6)

plt.tight_layout()

plt.show()
```

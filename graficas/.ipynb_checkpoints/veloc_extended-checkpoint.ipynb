```python
import numpy as np
import matplotlib.pyplot as plt

file_path = "/home/alumno23/MM/for-students/TAREA_PERFECTA/extended/veloc.xvg"
atom_name = "Cα-ARG-3"

def read_velocities(path):
    # Carga de todas las columnas: 0=tiempo, 1=vx, 2=vy, 3=vz, 4=|v|
    data = np.loadtxt(path, comments=["@", "#"])
    return data[:, 1], data[:, 2], data[:, 3]

try:
    vx, vy, vz = read_velocities(file_path)

    #Configuración de la figura
    fig, axs = plt.subplots(1, 3, figsize=(15, 6)) # Aumentamos un pelín el alto (de 5 a 6)
    
    #Título principal
    fig.suptitle(f"Análisis de Componentes de Velocidad de {atom_name}", 
                 fontweight='bold', fontsize=16, y=0.98)

    componentes = [vx, vy, vz]
    nombres = ["Vx", "Vy", "Vz"]
    colores = ["red", "purple", "cyan"]

    for i in range(3):
        axs[i].hist(componentes[i], bins=40, color=colores[i], alpha=0.7, edgecolor='black', lw=0.5)
        
        mean_v = np.mean(componentes[i])
        axs[i].axvline(mean_v, color='black', linestyle='--', lw=2, label=f'Media: {mean_v:.2f}')
        
        #Subtítulos de cada imagen
        axs[i].set_title(f"Componente {nombres[i]}", fontweight='bold', fontsize=12)
        
        axs[i].set_xlabel("Velocidad (nm/ps)")
        axs[i].set_ylabel("Frecuencia")
        axs[i].grid(alpha=0.2)

    #Ajustamos del diseño 
    plt.tight_layout(rect=[0, 0.03, 1, 0.95]) 
    
    plt.show()

except Exception as e:
    print(f"Error al procesar el archivo: {e}")
```

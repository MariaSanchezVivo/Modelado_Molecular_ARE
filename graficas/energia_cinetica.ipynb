```python
import numpy as np
import matplotlib.pyplot as plt

file_298 = "/home/alumno23/MM/for-students/TAREA_PERFECTA/4-analysis/analysis_298K/energy.xvg"
file_400 = "/home/alumno23/MM/for-students/TAREA_PERFECTA/4-analysis/analysis_400K/energy.xvg"

#Carga de datos optimizada (unpack=True para asignar directamente)
t_298, e_298 = np.loadtxt(file_298, comments=('@', '#'), unpack=True)
t_400, e_400 = np.loadtxt(file_400, comments=('@', '#'), unpack=True)

#Función para evitar repetir código
def plot_energy(time, energy, label, color, title):
    plt.figure(figsize=(9, 5))
    plt.plot(time, energy, label=label, color=color, lw=1.5)
    plt.title(title, fontweight='bold', fontsize=14)
    plt.xlabel("Tiempo (ps)")
    plt.ylabel("Energía Cinética (kJ/mol)")
    plt.legend()
    plt.grid(alpha=0.3)
    plt.tight_layout()
    plt.show()

#Gráficas
plot_energy(t_298, e_298, "298 K", "blue", "Energía Cinética a 298 K")
plot_energy(t_400, e_400, "400 K", "orange", "Energía Cinética a 400 K")

#Gráfica comparativa
plt.figure(figsize=(10, 6))
plt.plot(t_298, e_298, label="298 K", color="blue", lw=1.5, alpha=0.8)
plt.plot(t_400, e_400, label="400 K", color="orange", lw=1.5, alpha=0.8)

plt.title("Comparación de Energía Cinética", fontweight='bold', fontsize=14)
plt.xlabel("Tiempo (ps)")
plt.ylabel("Energía Cinética (kJ/mol)")
plt.legend()
plt.grid(alpha=0.3)
plt.tight_layout()

plt.show()
```
